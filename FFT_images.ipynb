```python
filterSize = 8
# get the path/directory
folder_dir = "/Users/gorkem.er/Documents/GitHub/A22_exp/images"
print(folder_dir)
for images in os.listdir(folder_dir):
 
    # check if the image ends with png
    if (images.endswith(".jpg")):
        print(images)
        img = Image.open("images/"+images).convert('L')

        dft = cv2.dft(np.float32(img),flags = cv2.DFT_COMPLEX_OUTPUT)
        # shift the zero-frequncy component to the center of the spectrum
        dft_shift = np.fft.fftshift(dft)

        rows = np.size(img, 0) #taking the size of the image
        cols = np.size(img, 1)
        crow, ccol = rows//2, cols//2

        # create a mask first, center square is 1, remaining all zeros
        mask = np.zeros((rows,cols,2),np.uint8)
        mask[crow-filterSize:crow+filterSize, ccol-filterSize:ccol+filterSize] = 1
        # apply mask and inverse DFT
        fshift2 = dft_shift*mask
        f_ishift2 = np.fft.ifftshift(fshift2)
        img_back2 = cv2.idft(f_ishift2)
        img_back2 = cv2.magnitude(img_back2[:,:,0],img_back2[:,:,1])

        plt.imshow(img_back2, cmap = 'gray')
        #cv2.imwrite("test.jpg",img_back2)
        imageio.imsave("HPF/"+"HPF_"+images, img_back2)
```
